```python
import os
import json
import numpy as np

data_folder = "/gscratch/xlab/hallisky/fast-detect-gpt/results/new_book"

master_float_list = []
labels = []

for file_name in os.listdir(data_folder):
    if file_name.endswith('.json'):
        file_path = os.path.join(data_folder, file_name)
        with open(file_path) as f:
            data = json.load(f)[:100]
            print(len(data))
            master_float_list.extend(data)
            if file_name == "new_book.json":
                labels.extend([0] * len(data))
            else:
                labels.extend([1] * len(data))

print("Master Float List:", master_float_list)
print("Labels:", np.unique(labels, return_counts=True))
```

```python
from sklearn.metrics import roc_auc_score

# Calculate AUC
auc = roc_auc_score(labels, master_float_list)
print(f"AUC: {auc}")
```
